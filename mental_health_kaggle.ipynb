## 2.1 -- Заполнение профессий

Если у нас студент, то профессия, напряженность на работе и удовлетворение раотой -- пустая
Иначе если у человека есть работа, то акажемическая нагрузка, CGPA и наслаждение работой -- пусто
Обработать пустые значения в dietary habits, degree, financial stress

```python
a = df.groupby('Working Professional or Student').size().reset_index().rename(columns = {0: 'Количество людей', 'Working Professional or Student': 'Профессия'})

f = px.pie(a, names = 'Профессия', values='Количество людей')
f.update_layout(title = {'text': 'Распределение профессий', 'x': 0.5}, legend = {'title': 'Профессии'})
f.show()
```

Из этого графика видно, что `nan`ы в последущих табличках обусловлены тем, что ~20% человек --- студенты
Теперь же заменим `nan`ы в табличках на более правдивые (на мой взгляд)

```python
display(df)
```

```python
def change_profession(row):
    if row['Working Professional or Student'] == 'Student':
        row['Profession'] = 'Student'
    return row['Profession']

df['Profession'] = df.apply(change_profession, axis = 1)

display(df)
```

```python
print(df.info())
```

Как видно, у нас многие значения содержат `NAN`, несмотря на то, что мы придумали адекватный фильтр. Попробуем тогда следующее: если у нас столбц `Work Pressure` не пустой, то это может говорить нам о том, что человек **явно не студент и у него есть работа**
<br>
Но сначала давайте просто посмотрим, какие вообще у нас есть профессии, чтобы в случае чего заполнить проппуски подходящими

```python
ac = df.groupby('Profession').size().reset_index().rename(columns = {0: 'Количество людей'}).sort_values('Количество людей', ascending=False)

ac['col'] = 'blue'
ac['col'].iloc[0] = 'red'

# Я хочу, чтобы у нас случайно выбирались профессии с определенными вероятностями, чтобы их соотственно заполнять в пустые ячейки

ac['prob'] = ac['Количество людей'] / ac['Количество людей'].sum()

display(ac[:20])

f = px.bar(ac[:20], x = 'Profession', y = 'Количество людей', labels={'Profession':  'Профессия'}, color = 'col', color_discrete_map={"red": 'red', 'blue': 'blue'})
f.update_layout(title = {'text': 'Самые популярные профессии (топ 20)', 'x' : 0.5}, showlegend = False)

f.show()
```

```python
# Посмотрим, что у нас есть по оставшимся нулевым профессиям 
ab = df[
    pd.isna(df['Profession'])
]
display(ab)
```

Когда мы строили график с профессиями, я добавил еще колоночку вероятностей для каждой профессии, теперь, зная эти вероятности при помощи них  мы будем заполнять наши пропущенные значения

```python
def assign_profession(row):
    if pd.isna(row['Profession']):
        row['Profession'] = np.random.choice(ac['Profession'], p = ac['prob'])
    elif pd.isna(row['Work Pressure']) and not pd.isna(row['Academic Pressure']) and pd.isna(row['Profession']):
        row['Profession'] = 'Student'
    else:
        row['Profession'] = row['Profession']
    return row['Profession']

df['Profession'] = df.apply(assign_profession, axis = 1)

display(df)
```

```python
print(df.info())
```

## 2.2 -- Заполенение академической нагрузки

```python
def replace_academic_pressure(row):
    if row['Profession'] != 'Student' and pd.isna(row['Academic Pressure']):
        row['Academic Pressure'] = 0
    return row['Academic Pressure']

df['Academic Pressure'] = df.apply(replace_academic_pressure, axis = 1)

# Найдем среднюю академическую загруженность студентов

c = df[
    (df['Profession'] == 'Student') & ~(pd.isna(df['Academic Pressure']))
]

pressure_mean = c['Academic Pressure'].mean()
print(pressure_mean)

df['Academic Pressure'] = df['Academic Pressure'].fillna(pressure_mean)

display(df)
```

```python
print(df.info())
```

## 2.3 -- Заполнение рабочей загрузки

```python
def replace_work_pressure(row):
    if row['Profession'] == 'Student':
        row['Work Pressure'] = 0
    return row['Work Pressure']

df['Work Pressure'] = df.apply(replace_work_pressure, axis = 1)

display(df)
```

```python
print(df.info())
```

```python
a =  df[
    pd.isna(df['Work Pressure'])
]
display(a)
```

Нам ничего не остается, кроме как заполнить оставшиеся пропуски средними значениями

```python
w_mean = df['Work Pressure'].mean()
print(w_mean)

df['Work Pressure'] = df['Work Pressure'].fillna(w_mean)

display(df)
```

```python
print(df.info())
```

## 2.4 -- Заполнение CGPA (оценки)

```python
display(df)
```

Сначала проверим, что действительно ли у всех работающих лбдей нету CGPA, чтобы было дальше нам проще

```python
a = df.groupby('Working Professional or Student')['CGPA'].mean().reset_index().rename(columns={0: 'Средний балл CGPA'}).sort_values('CGPA', ascending = False)
print(a)

avg_grade = a.iloc[0, 1]
avg_grade_st = a.iloc[0, 1]
print(avg_grade) # средняя оценка работающих людей
print(avg_grade_st) # средняя оценка студентов

f = px.bar(a, 
           x = 'Working Professional or Student', 
           y = 'CGPA', 
           color = 'Working Professional or Student',
           labels = {'Working Professional or Student': 'Род деятельности'}, 
           hover_name='Working Professional or Student')
f.update_layout(title = {'text':  'Средний балл CGPA работающих людей и студентов', 'x': 0.5}, showlegend = False)
f.show()
```

Ага, что-то есть, даже балл больше, чем у студентов, а теперь посмотрим на наших ребят

```python
stu = df[
    (df['Working Professional or Student'] == 'Working Professional') & ~(pd.isna(df['CGPA']))
]
st = stu['CGPA'].std() # стандартное отклонение для наших данных
print(st)
display(stu)
```

Понятно, это просто студенты, которые работают и совмещают учебу
<br>
Думаю, можно сделать вывод, что следует также сделать и для остальных работающих людей, возможно, стоит распределить их баллы по нормальному распределению
<br>
В таком случае, нам надо рассчитать дисперсию или же стандартное отклонение в квадрате $\sigma^2$ 
<br>
Среднее значение у нас уже есть

```python
working_val = np.random.normal(loc = avg_grade, scale=st)
print(working_val)
```

Стоит также проверить, что у студентов нету пропусков в баллах

```python
display(df[
    (pd.isna(df['CGPA'])) & (df['Working Professional or Student'] == 'Student')
])
```

Как видно, некоторые пропуски тоже имеются, поэтому попробуем заполнить значениями из нормального распределения и для этих студентов тоже
<br>
Сначала построим график распределения существующих оценок

```python
stud = df[
    ~(pd.isna(df['CGPA'])) & (df['Working Professional or Student'] == 'Student')
]
f = px.histogram(stud, x = 'CGPA', nbins = 30)
f.update_layout(title = {'text': 'Распределение оценок между учениками', 'x': 0.5}, yaxis_title = 'Количество учеников')
f.show()
```

```python
stud_std = stud['CGPA'].std()

student_val = np.random.normal(loc = avg_grade_st, scale=stud_std)
print(student_val)
```

Теперь заполним пропущенные значения значениями из нормальных распределений

```python
def replace_nan_CGPA(row):
    if row['Working Professional or Student'] == 'Student' and pd.isna(row['CGPA']):
        row['CGPA'] = student_val
    elif row['Working Professional or Student'] != 'Student' and pd.isna(row['CGPA']):
        row['CGPA'] = working_val
    else:
        row['CGPA'] = row['CGPA']
    return row['CGPA']

df['CGPA'] = df.apply(replace_nan_CGPA, axis = 1)
display(df)
```

```python
print(df.info())
```
